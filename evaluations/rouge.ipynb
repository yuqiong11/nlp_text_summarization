```python
from datasets import load_metric

# read all generated and reference summaries from external
gen = pd.read_csv('gen_summary_gpt2.csv', header=0)['gensummary']
ref = pd.read_csv('ref_summary_gpt2.csv', header=0)['refsummary']
# or get directly from dataframe
# ref = test_sum

# open a new .csv, loop over all generated summaries, write the rouge1|rouge2|rougeL scores into .csv
with open('rouge_gpt2.csv', 'w', newline='') as csvfile:
  fieldnames = ['rouge1', 'rouge2', 'rougeL']
  writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
  writer.writeheader()
  for r, g in zip(ref, gen):
      rouge = load_metric('rouge')
      rouge.add(prediction=g,reference=r)
      rouge_ = rouge.compute()
      # only consider recall score
      writer.writerow({'rouge1':rouge_['rouge1'].mid.recall, 'rouge2':rouge_['rouge2'].mid.recall, 'rougeL':rouge_['rougeL'].mid.recall})
```

```python
# get the average rouge1|rouge2|rouge3|rougeL scores 
rouge_avg = load_metric('rouge')

for r, g in zip(ref, gen):
  rouge_avg.add(prediction=g,reference=r)
score = rouge_avg.compute(rouge_types=['rouge1',"rouge2","rouge3",'rougeL'])

for i, j in score.items():
  print(i, j.mid.recall)
```
